##### Notes   
RANSAC (RANdom SAmple Consensus) algorithm.  
It is an iterative algorithm for robust estimation of parameters from a subset of inliers from the complete data set.

RANSAC Regression is non-deterministic, iterative, and robust to outliers.
It works as follows:
- Select a subset of data
- Fit a model
- Calculate residuals
- Classify points as outliers/inliers based on thresholds applied to residuals
- Repeat (until max iterations or when a condition is met)

[Read more](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.RANSACRegressor.html)

````python
sns.distplot(df2['Height_cm'])
````

````python
sns.distplot(df['Weight_Kg'])
````

````python
# removing outliers using interquartile range approach (IQR)
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
````

````python
# function to remove outliers
def outliers(column):
  sorted(column)
  Q1, Q3 = np.percentile(column, [25, 75])
  IQR = Q3 - Q1
  lower_range = Q1 - (1.5 * IQR)
  upper_range = Q3 + (1.5 * IQR)
  return lower_range, upper_range
````

````python
import numpy as np
````

````python
lower_weight, upper_weight = outliers(df['Weight_Kg'])
````

````python
lower_height, upper_height = outliers(df['Height_cm'])
````

````python
df[(df.Height_cm < lower_height) | (df.Height_cm > upper_height)]
````
